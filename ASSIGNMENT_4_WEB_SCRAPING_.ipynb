```python
### import all required libraries
import pandas as pd
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
```

```python
### Load webdriver
driver = webdriver.Chrome('chromedriver.exe')
```

## **1) Scrape the details of most viewed videos on YouTube from Wikipedia**

```python
### Opening the wikipedia
driver.get('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos')
```

```python
### Get rows from the table "Top 30 most-viewed YouTube videos"
table_rows = driver.find_element_by_class_name('wikitable').find_elements_by_xpath("tbody/tr")
```

```python
### Create empty list for storing data 
rank = []
name = []
artist = []
upload_date = []
views = []
```

```python
### Loop to scrape values from each rows
for i in table_rows:
    row_data = i.find_elements_by_tag_name('td') # Get row data elements
    rank.append(row_data[0].text.strip()[:-1]) # Scrape rank
    name.append(row_data[1].text.split('\"')[1]) # Scrape name
    artist.append(row_data[2].text) # scrape artist name
    views.append(row_data[3].text) # Scrape views
    upload_date.append(row_data[4].text) # Scrape video upload date
```

```python
### Create dictionary of the scraped data
data_dict = dict(
    rank=rank,
    name=name,
    artist=artist,
    upload_date=upload_date,
    views_in_billions=views)

### Create data frame of the dictionary data and save as csv file
data = pd.DataFrame(data_dict)
data.to_csv('Top 30 most-viewed YouTube videos.csv')
data
```

## **2) Scrape the details team India’s international fixtures from bcci.tv**

```python
### Opening the bcci.tv fixtures page
driver.get('https://www.bcci.tv/')

### Click on INTERNATIONAL drop down button
WebDriverWait(driver, 10).until(EC.element_to_be_clickable(
    (By.XPATH,"//li[@class='navigation__item  ']"))).click() # Wait until the 'International' drop down button is clickable

### Click on fixtures
WebDriverWait(driver, 10).until(EC.element_to_be_clickable(
    (By.XPATH,"//a[contains(text(),'Fixtures')]"))).click() # Wait until the 'Fixtures' in drop down list is clickable
```

```python
### Get list of fixtures
event_list = driver.find_elements_by_xpath("//a[contains(@class,'js-match')]")
```

```python
### Create empty list for storing data 
match_title = []
series = []
place = []
date = []
time = []
```

```python
### Loop to scrape data from each fixture
for match in event_list:
    
    ### Series
    series.append(match.find_element_by_class_name('fixture__format-strip').text.replace('\n',' '))
    
    ### Match Title
    match_title.append(match.find_element_by_class_name('fixture__additional-info').find_element_by_tag_name('strong').text)
    
    ### Place
    place.append(match.find_element_by_class_name('fixture__additional-info').find_element_by_tag_name('span').text)
    
    ### Find date and time element
    date_time = match.find_element_by_class_name('fixture__datetime').text.split('\n')
    date.append(' '.join(date_time[:-1])) # Date
    time.append(date_time[-1]) # Time
```

```python
### Create dictionary of the scraped data
data_dict = dict(
    match_title=match_title,
    series=series,
    place=place,
    date=date,
    time=time)

### Create data frame of the dictionary data and save as csv file
data = pd.DataFrame(data_dict)
data.to_csv('BCCI International fictures.csv')
data
```

## **5) Scrape the details of trending repositories on Github.com**

```python
### Opening the github page
driver.get('https://github.com/')

### Click on Explore nav button
WebDriverWait(driver, 10).until(EC.element_to_be_clickable(
    (By.XPATH,"//summary[contains(text(),'Explore')]"))).click() # Wait until the link is clickable

### Click on Trending in the navigation list
WebDriverWait(driver, 10).until(EC.element_to_be_clickable(
    (By.XPATH,"//a[contains(text(),'Trending')]"))).click() # Wait until the link is clickable
```

```python
### Find repository elements 
articles = driver.find_elements_by_class_name('Box-row')
```

```python
### Create empty list for storing data
rep_title = []
rep_desc = []
contrib_cnt = []
lang = []
```

```python
### Loop over each trending articles or repository
for repo in articles:
    
    ### Scrape the title
    rep_title.append(repo.find_element_by_class_name('h3').text.split('/')[1].strip())
    
    ### Scrape description and append '-' if description not available
    try:
        rep_desc.append(repo.find_element_by_tag_name('p').text)
    except NoSuchElementException:
        rep_desc.append('-')
        
    ### Scrape programming language and append '-' if language not available
    try:
        lang.append(repo.find_element_by_css_selector("span[itemprop='programmingLanguage']").text)
    except NoSuchElementException:
        lang.append('-')
        
    ### Count the number of contibutors 
    contrib_cnt.append(len(repo.find_elements_by_css_selector("span[class='d-inline-block mr-3'] a")))
```

```python
### Create dictionary of the scraped data
data_dict = dict(
    rep_title=rep_title,
    rep_desc=rep_desc,
    contrib_cnt=contrib_cnt,
    lang=lang)

### Create data frame of the dictionary data and save as csv file
data = pd.DataFrame(data_dict)
data.to_csv('github trending repositories.csv')
data
```

## **6) Scrape the details of top 100 songs on billiboard.com**

```python
### Opening the billboard page
driver.get('https://www.billboard.com/')

### Click on HOT 100
WebDriverWait(driver, 10).until(EC.element_to_be_clickable(
    (By.XPATH,"//a[contains(@class,'header__subnav__link') and contains(text(),'Hot 100')]"))).click() # Wait until the link is clickable
```

```python
### Find rows or elements containing song details
hot100 = driver.find_elements_by_class_name('chart-list__element')
```

```python
### Create empty list for storing data
song_name = []
artist_name = []
last_week_rank = []
peak_rank = []
weeks_on_chart = []
```

```python
### Loop to iterate over songs
for song in hot100:
    
    ### Find song name
    song_name.append(song.find_element_by_class_name('chart-element__information__song').text)
    
    ### Find artist name
    artist_name.append(song.find_element_by_class_name('chart-element__information__artist').text)
    
    ### Find element having all the rank information
    meta = song.find_elements_by_css_selector("div.chart-element__meta")
    last_week_rank.append(meta[-3].text) # Last week rank
    peak_rank.append(meta[-2].text) # Peak rank
    weeks_on_chart.append(meta[-1].text) # Weeks on chart
```

```python
### Create dictionary of the scraped data
data_dict = dict(
    song_name=song_name,
    artist_name=artist_name,
    last_week_rank=last_week_rank,
    peak_rank=peak_rank,
    weeks_on_chart=weeks_on_chart)

### Create data frame of the dictionary data and save as csv file
data = pd.DataFrame(data_dict)
data.to_csv('billboard hot 100.csv')
data
```

## **7) Scrape the details of Data science recruiters from naukri.com**

```python
### Opening the naukri page
driver.get('https://www.naukri.com/')

### Select the Recruiters nav element
WebDriverWait(driver, 10).until(EC.element_to_be_clickable(
    (By.XPATH,"//a[contains(@title,'Search Recruiters')]"))).click() # Wait until the element can be selected

### Get the window handle of new window
child_window_handle = driver.window_handles[1]

### Swtich control to the new child window
driver.switch_to.window(child_window_handle)
```

```python
### Find the elements of search field
skill_field = driver.find_element_by_xpath("//div[@id='skill']/div/div/input") # Skill field
search_button = driver.find_element_by_id('qsbFormBtn') # Search button

### Search for 'Data Science'
skill_field.clear() # Clear the field
skill_field.send_keys('Data Science') # Insert Data Science as input
search_button.click() # Click send
```

```python
### Create empty list for storing data
name = []
desig = []
company = []
skill = []
loc = []
```

```python
### Initializing page count
count = 0

### Set flag for page limit
page_limit = 5 # The scraping should stop after 5th page. To scrape all available details, set it to a large number

### Loop over each page
while count < page_limit:

    ### Find elements containg the recruiters details
    recruiters_cards = WebDriverWait(driver, 20).until(EC.presence_of_all_elements_located((By.CLASS_NAME,"recSec")))

    ### Loop to iterate over each card or recruiter details
    for rec in recruiters_cards:

        ### Find the element having name, designation and company
        details = rec.find_elements_by_tag_name('span')[:3]
        name.append(details[0].text) # Name
        desig.append(details[1].text) # Designation
        company.append(details[2].text) # Company Name

        ### Get the location of the recruiter
        try:
            loc.append(rec.find_element_by_css_selector('small.ellipsis').text)
        except NoSuchElementException:
            loc.append('-') # Insert '-' if location not found

        ### Get the description of skills hiring for
        try:
            skill.append(rec.find_element_by_css_selector('div.hireSec').text)
        except NoSuchElementException:
            skill.append('-') # Insert '-' if description not found
            
    ### Increment count variable
    count = count + 1
        
    ### Break when count reaches the page limit
    if count == page_limit:
        break
    
    ### Click on next page
    try:
        driver.find_element_by_xpath("//div[@class='pagination']/button[text()='Next']").click()
    except NoSuchElementException:
        break # Break if next page dont exist
```

```python
### Create dictionary of the scraped data
data_dict = dict(
    name=name,
    designation=desig,
    company=company,
    skill_hiring_for=skill,
    loc=loc)

### Create data frame of the dictionary data and save as csv file
data = pd.DataFrame(data_dict)
data.to_csv('naukri recruiters.csv')
data
```

## **10) Details of Datasets from UCI machine learning repositories**

```python
### Opening the UCI machine learning page
driver.get('https://archive.ics.uci.edu/')

### Click on view all dataset
driver.find_element_by_xpath("//a[text()='view all data sets']").click()
```

```python
### Find the row elements of the dataset table
datasets = driver.find_elements_by_xpath("//table[@border='1' and @cellpadding='5']/tbody/tr")[1:]
```

```python
### Create empty lists for storing data
dataset_name = []
data_type = []
task = []
attribute_type = []
instances = []
attributes = []
year = []
```

```python
### Loop over each dataset
for dataset in datasets:
    
    ### Get the details of each dataset
    details = dataset.find_elements_by_css_selector("tr > td")[2:]
    fill_empty = lambda x: '-' if x.text.strip() == '' else x.text.strip() # Function to replace empty values with '-'
    details = [fill_empty(i) for i in details] # Replace all the empty values
    
    ### Append data to the lists
    dataset_name.append(details[0]) # Name of the dataset
    data_type.append(details[1]) # Data type
    task.append(details[2]) # Default task type
    attribute_type.append(details[3]) # Types of attributes
    instances.append(details[4]) # Number of instances
    attributes.append(details[5]) # Number of attributes
    year.append(details[6]) #year
```

```python
### Create dictionary of the scraped data
data_dict = dict(
    dataset_name=dataset_name,
    data_type=data_type,
    default_task=task,
    attribute_types=attribute_type,
    no_of_instances=instances,
    no_of_attributes=attributes,
    year=year)

### Create data frame of the dictionary data and save as csv file
data = pd.DataFrame(data_dict)
data.to_csv('UCI Machine learning dataset details.csv')
data
```
