### 1) To display all the header tags from ‘en.wikipedia.org/wiki/Main_Pag'.

```python
def get_wiki_main_page(url):
    
    """
    Finds all the header tags
    """
    
    ### Fetch HTML text
    html_text = get_html_doc(url)
    
    ### Use html parser from beautiful soup library to  parse text into HTML 
    bs = BeautifulSoup(html_doc.text, 'html.parser')
    
    ### Find all headers tags like header, h1, h2, h3, h4, h5 and h6
    header_tags = bs.find_all(['header','h1','h2','h3','h4','h5', 'h6'])
    
    return header_tags
```

### 2) To display IMDB’s Top rated 100 movies data

```python
def get_IMDB_top_100_movies(url):
    
    """
    Scrapes top 100 IMDB movies
    """
    
    Name = []
    IMDB_rating = []
    Years = []
        
    for i in url:
        imdb_html_text = get_html_doc(i)
        
        ### Use html parser from beautiful soup library to  parse text into HTML 
        bs = BeautifulSoup(imdb_html_text, 'html.parser')

        ### Movie names
        for j in bs.find_all('h3',class_='lister-item-header'):
            Name.append(j.a.text)
            
        ### Rating
        for j in bs.find_all('div',class_='inline-block ratings-imdb-rating'):
            IMDB_rating.append(j['data-value'])

        ### Year of release
        for j in bs.find_all('span',class_='lister-item-year text-muted unbold'):
            Years.append(j.text[1:-1])
        
    movies = list(zip(Name, IMDB_rating, Years))
    return movies
```

```python
### List of movie urls (Page 1 and page 2)
url = ['https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc',
      'https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc&start=51&ref_=adv_nxt']
imdb_data = get_IMDB_top_100_movies(url)
```

```python
### Converting to CSV File
imdb_df = pd.DataFrame(imdb_data, columns = ['Movie_name','IMDB_rating','Year_of_release'])
imdb_df.to_csv('IMDB_Top_100_movies.csv')
```

### 3) To display IMDB’s Top rated 100 Indian movies data

```python
### Function to scrape top 100 indian movies in IMDB
def get_IMDB_top_100_indian_movies(url):
    
    """
    Scrapes top 100 indian movies
    """
    
    Name = []
    IMDB_rating = []
    Years = []
    
    ### Fetching HTML
    imdb_html_text = get_html_doc(url)
    
    ### Use html parser from beautiful soup library to  parse text into HTML 
    bs = BeautifulSoup(imdb_html_text, 'html.parser')
    
    ### Get tags having movie names and year of release
    name_year_tags = bs.find_all('h3', class_='lister-item-header')
    
    ### Get rating of the movie
    rating_tags = bs.find_all('div', class_='ipl-rating-star small')

    for i in range(0,100):
        Name.append(name_year_tags[i].a.text)
        Years.append(name_year_tags[i].find(class_='lister-item-year text-muted unbold').text[1:-1])
        IMDB_rating.append(rating_tags[i].find(class_="ipl-rating-star__rating").text)
        
    movies = list(zip(Name, IMDB_rating, Years))
    return movies
```

```python
url = 'https://www.imdb.com/list/ls009997493/'
indian_imdb_data = get_IMDB_top_100_indian_movies(url)
```

```python
### Converting to CSV file
indian_imdb_df = pd.DataFrame(indian_imdb_data, columns = ['Movie_name','IMDB_rating','Year_of_release'])
indian_imdb_df.to_csv('IMDB_Top_100_indian_movies.csv')
```

### 4) To scrap book name, author name, genre and book review of any 5 books from 'www.bookpage.com'

```python
def get_book_reviews(url):

    """
    Scrapes book review
    """
    
    book_html_text = get_html_doc(url)
    
    ### Use html parser from beautiful soup library to  parse text into HTML 
    bs = BeautifulSoup(book_html_text, 'html.parser')

    ### Get tags with books information
    books_tags = bs.find_all('div', class_='bp-block article-info')[:5]
    
    for i in range(0,5):
        print('Name of Book : ',' '.join(re.findall('(\w+)',books_tags[i].find('h4',).a.text)))
        print('Author Name : ',books_tags[i].find('p',class_='sans bold').text.strip())
        print('Genre : ',books_tags[i].find('p',class_='genre-links hidden-phone').text.replace('\n','').replace('/',','))
        
        ### Fetching book reviews
        book_review = BeautifulSoup(
            get_html_doc('https://bookpage.com'+books_tags[i].find('div',class_='read-full').a.attrs['href']),
            'html.parser')
        print('Review : ')
        print(book_review.find('div',class_='article-body').p.text)
        print('---------------------------------------------')
```

### Functions to scrape cricket rankings of cricket from 'www.icc-cricket.com'

```python
def get_top_10_team_cricket(url):
    
    """
    Scarpes top 10 ODI teams in cricket along with the records for matches, points and rating.
    
    """
    
    ### Fetching HTML
    odi_html_text = get_html_doc(url)
    bs = BeautifulSoup(odi_html_text, 'html.parser')
    
    ### Find the table tags
    table_tags = bs.find('tbody').find_all('tr')
    team,matches,points,rating = [],[],[],[]
    
    for i in range(0,10):
        ### Get team details
        table_row = table_tags[i].find_all('td')
        team.append(table_row[1].text.strip().split('\n')[1])
        matches.append(table_row[2].text.strip())
        points.append(table_row[3].text.strip())
        rating.append(table_row[4].text.strip())
    
    top_10_ODI_teams = dict(team=team,matches=matches,points=points,rating=rating)
    return top_10_ODI_teams
```

```python
def get_top_10_players(url):
    
    """
    Scrapes top 10 players in cricket
    """
    
    ### Fetching HTML tags
    odi_html_text = get_html_doc(url)
    bs = BeautifulSoup(odi_html_text, 'html.parser')

    ### Find table tags
    table_tags = bs.find_all('tr')
    player,team,rating = [],[],[]
    
    for i in range(1,11):
        ### Get players details
        table_row = table_tags[i].find_all('td')
        player.append(table_row[1].text.strip())
        team.append(table_row[2].text.strip())
        rating.append(table_row[3].text.strip())

    top_players = dict(player=player,team=team,rating=rating)
    return top_players
```

### 7) To scrape details of all the mobile phones under Rs. 20,000 listed on Amazon.in

```python
def extract_amazon_products_mobile(url=None):

    """
    Scrapes product details of mobile phones from amazon india website
    """
    
    product_name,price,image_url,average_rating=[],[],[],[]
    
    HEADERS = ({'User-Agent':
            'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/44.0.2403.157 Safari/537.36'})

    ### Iterate over pages
    for page in ind:

        ### Fethcing HTML
        url = 'https://www.amazon.in/s?i=electronics&bbn=1389401031&rh=n%3A976419031%2Cn%3A1389401031%2Cn%3A1389432031%2Cp_36%3A-2000000&dc&page={0}&qid=1612459959&rnid=1318502031&ref=sr_pg_{0}'.format(page)
        html = requests.get(url,headers=HEADERS)
        bs = BeautifulSoup(html.text, 'html.parser')

        ### Div tags having product details
        div_tag = bs.find_all('div','s-result-item s-asin sg-col-0-of-12 sg-col-16-of-20 sg-col sg-col-12-of-16')
        
        ### Show pages
        print(page,end='|')
        
        
        ### Iterate over product lists
        for i in div_tag:

            ### Get details of mobile
            name = i.find('a',class_='a-link-normal a-text-normal')
            rat = i.find('span',class_='a-icon-alt')
            im = i.find('img',class_='s-image')
            pr = i.find('span',attrs={'class':'a-price','data-a-size':'l'})
            
            ### Replace empty entries with None 

            if name != None:
                product_name.append(name.text.strip())
            else:
                product_name.append('None')
            if rat != None:
                average_rating.append(rat.text.strip().split(' ')[0])
            else:
                average_rating.append('None')
            if im != None:
                image_url.append(im.attrs['src'])
            else:
                image_url.append('None')
            if pr != None:
                price.append(pr.span.text)
            else:
                price.append('None')
                
        ### Induce delay for mimicing manual action
        time.sleep(random.randint(5,10))
        
        ### Convert to dict
        data_mobile =  dict(product_name=product_name,
                 price=price,
                 image_url=image_url,
                 average_rating=average_rating)
        return data_mobile
```

### 8)  To extract information about the local weather from the National Weather Service website of USA, https://www.weather.gov/ for the city, San Francisco

```python
def get_weather_forecast(url):
    
    """
    Scrapes 7 day extended forecast display for the city San Fransisco
    """
    
    weather_forecast = get_html_doc(url)
    bs = BeautifulSoup(weather_forecast, 'html.parser')

    seven_day_forecast = bs.find_all('li',class_='forecast-tombstone')
    det_forecast = bs.find_all('div',class_='row-forecast')
    period, short_desc, temp, desc = [],[],[],[]

    for i in seven_day_forecast:
        period.append(i.find('p',class_='period-name').text)
        short_desc.append(i.find('p',class_='short-desc').text)
        temp.append(i.find('p',class_='temp').text)

    for i in det_forecast:
        desc.append(i.find('div','col-sm-10 forecast-text').text.strip())
        
    return list(zip(period, short_desc, temp, desc))
```

```python
### URL for weather in San Fransisco
url = 'https://forecast.weather.gov/MapClick.php?lat=37.777120000000025&lon=-122.41963999999996#.YB62DegzbDd'
pd.DataFrame(get_weather_forecast(url),columns=['period','short_desc','temp','desc'])
```

### 9) To scrape ‘software developer’ job listings from ‘Monster.com’

```python
def get_job_listings(url,page_limit):
    
    """
    Function to scrape job listings from Monster.com
    """
    
    HEADERS = ({'User-Agent':
            'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/44.0.2403.157 Safari/537.36'})

    job_title, company, location, experience, salary,desc,=[],[],[],[],[],[]

    count = page_limit
    while count:

        print('page ',page_limit+1-count)

        html = requests.get(url,headers=HEADERS)
        bs = BeautifulSoup(html.text, 'lxml')
        job_list = bs.find('div',id='srp-jobList')
        job = job_list.find_all('div','card-panel apply-panel job-apply-card')

        for i in job:

            job_title.append(i.find('div',class_='job-tittle').h3.a.text.strip())
            company.append(i.find('span',class_='company-name').text.strip())
            loc = i.find_all('span',class_='loc')
            location.append(', '.join(re.findall('([a-zA-Z]+)',loc[0].small.text)))
            experience.append(loc[1].small.text.strip())
            salary.append(loc[2].small.text.strip())
            desc.append(i.find('p',class_='job-descrip').text.strip())

        url = str(bs.find('div',id='pag-data',class_='srp-navigation').a.attrs['href'])
        
        count = count - 1

        time.sleep(random.randint(5,10))
        
    lists = dict(job_title=job_title,company=company,
                location=location,experience=experience,
                salary=salary,desc=desc)

    return lists
```

```python
### URL for software developer jobs
url = 'https://www.monsterindia.com/search/software-developer-jobs'
data = pd.DataFrame(get_job_listings(url,5))
```
